```python
# Dependencies
import pandas as pd
import requests
import json
import xmltodict
from pprint import pprint


# Zillow Key
from config import zkey
```

```python
# Collect Zillow data - Austin neighborhoods overview
baseURL = 'http://www.zillow.com/webservice/GetRegionChildren.htm'
params = {
    'zws-id': zkey,
    'state': 'tx',
    'city': 'austin',
    'childtype': 'neighborhood'
}

response = requests.get(baseURL, params=params)
```

```python
# Use xmltodict to convert the response from XML to JSON
cleanResponse = pprint(json.dumps(xmltodict.parse(response.content), sort_keys=True))
```
